```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def fx(x):
    return 3*x**2-3*x+4

def derv(x):
    return 6*x-3
```

```python
#plot functions

x = np.linspace(-2,2,2001)
plt.plot(x,fx(x),x,derv(x))

plt.show()
```

```python
localmin = np.random.choice(x,1)
print(localmin)
learning_rate =.01
epos =10000

for i in range(epos):
    grad = derv(localmin)
    #print(grad)
    localmin = localmin-learning_rate*grad
print(localmin)
```
